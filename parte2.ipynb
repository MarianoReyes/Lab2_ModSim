## 1.1 
Sí, este evento puede ser modelado por una distribución de Poisson. La razón es que la distribución de Poisson se usa para modelar el número de veces que un evento ocurrió en un intervalo de tiempo. En este caso, el evento es la llegada de un bus a la parada y el intervalo de tiempo es 30 minutos.
## 1.2

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Definir la tasa de llegada promedio
lmbda = 2

# Crear una lista con los números de buses (desde 0 hasta 100)
bus_counts = list(range(0, 101))

# Calcular las probabilidades para cada número de buses
probs = [poisson.pmf(k, lmbda) for k in bus_counts]

# Graficar las probabilidades
plt.figure(figsize=(12,6))
plt.bar(bus_counts, probs)
plt.xlabel('Número de buses')
plt.ylabel('Probabilidad')
plt.title('Distribución de Poisson para el número de buses que llegan en 30 minutos')
plt.show()


```

## 2.2  Graficar la CDF de la distribución exponencial 

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 1000)
y = expon.cdf(x, scale=1/mu)

plt.plot(x, y)
plt.title('CDF de la distribución exponencial')
plt.xlabel('Tiempo intermedio')
plt.ylabel('CDF')
plt.show()
```

## 2.3  Graficar los tiempos intermedios para los primeros 500 pacientes

```python
import numpy as np
import matplotlib.pyplot as plt
def gen_tiempos(promedio_llegadas,pacientes):
    tiempos_intermedios = np.random.exponential(1/promedio_llegadas, pacientes)
    tiempos_acumulados = np.cumsum(tiempos_intermedios)
    return tiempos_acumulados

llegadas =5
pacientes = 500

tiempo_intermedio = gen_tiempos(llegadas,pacientes)

plt.plot(range(1,pacientes+1),tiempo_intermedio)
plt.title('Tiempos Intermedios para lso 500 primeros pacientes')
plt.xlabel('Pacientes')
plt.ylabel('Tiempo intermedio')
plt.grid(True)
plt.show()

```

#### Qué forma tiene la gráfica?

La grafica tiene una tendencia lineal con una pendiente positiva, a memdida de que mas pacienetes llegan a la clinica, el tiempo de espera aumenta.

#### ¿Cuál es la relación que se observa entre esta y la gráfica del punto anterior?

Tenemos que la grafica del punto anterior es una grafica exponencial, mientras que la grafica de este punto es una grafica lineal, por lo que no tienen o dependen de la misma funcion.

## 3.1 T1 y T2 hasta Tk son variables aleatorias?

Sí, las variables T1, T2, ..., Tk son variables aleatorias. Cada una de estas variables representa un punto en el tiempo en el que llega un paciente. Dado que los tiempos intermedios entre las llegadas de los pacientes (X1, X2, ..., Xk) son variables aleatorias exponenciales independientes, la suma de estas también es una variable aleatoria. Esto se debe a que el resultado de la suma depende de las realizaciones de las variables aleatorias exponenciales.

## 3.2 Siimulacion Poisson

```python
# Generar tiempos intermedios para 100 pacientes
interarrival_times_100 = np.random.exponential(1/mu, 100)

# Calcular los tiempos de llegada acumulativos
arrival_times = np.cumsum(interarrival_times_100)

# Crear una gráfica
plt.step(arrival_times, np.arange(1, 101), where='post')
plt.title('Proceso de Poisson para las llegadas de los pacientes')
plt.xlabel('Tiempo')
plt.ylabel('Número de pacientes')
plt.show()
```
