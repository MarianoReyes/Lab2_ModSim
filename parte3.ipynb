```python
# 1.

import numpy as np

p = 0.2
num_horas = 10
num_clientes = np.random.geometric(p, num_horas)

print(num_clientes)
```

```python
# 2.

import matplotlib.pyplot as plt

arrival_times = np.cumsum(num_clientes)  # Calcular los tiempos acumulativos de llegada
time_intervals = np.diff(arrival_times)  # Calcular los intervalos de tiempo entre llegadas

plt.plot(arrival_times[1:], time_intervals, marker='o')
plt.xlabel('Tiempo acumulado de llegada')
plt.ylabel('Intervalo de tiempo entre llegadas')
plt.title('Patrón de llegada de clientes')
plt.show()

average_time_between_arrivals = np.mean(time_intervals)
print("Tiempo promedio entre llegadas:", average_time_between_arrivals)
```

```python
# 1.

import numpy as np

r = 5
p = 0.4
sample = 500

service_times = np.random.negative_binomial(r, p, sample)

print(service_times)
```

```python
# 2.

import matplotlib.pyplot as plt

plt.hist(service_times, bins=30)
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.title('Histograma de tiempos de servicio')
plt.show()

media = np.mean(service_times)
varianza = np.var(service_times)

print("Media:", media)
print("Varianza:", varianza)
```

3. Por la forma teórica de la distribución binomial negativa, podemos ver que nuestro histograma si se ajusta a dicha distribución.

```python
# Cálculo de la media y la varianza teóricas
media_teorica = r * (1 - p) / p
varianza_teorica = r * (1 - p) / (p**2)

print(media_teorica)
print(varianza_teorica)
```

4. La media teórica de una distribución binomial negativa con parámetros r y p es dada por:
    μ = r * (1 - p) / p

    La varianza teórica de la distribución binomial negativa es:
    σ^2 = r * (1 - p) / (p^2)

    Con los resultados obtenidos teóricos podemos verificar que efectivamente nuestro histograma se ajusta correctamente a la distribución binomial negativa.

## Task 4

```python
# 1. 

import numpy as np

# Generación de la muestra aleatoria de datos
data = np.random.rand(1000)

print(data)
```

```python
# 2. 

from scipy.stats import norm, expon

# Ajuste de distribución normal
params_normal = norm.fit(data)
mean_normal, std_normal = params_normal

# Ajuste de distribución exponencial
params_exponential = expon.fit(data)
scale_exponential = params_exponential[1]

print("Parámetros ajustados para la distribución normal: ", mean_normal, std_normal)
print("Parámetros ajustados para la distribución exponencial: ", scale_exponential)
```

```python
# 3.

import matplotlib.pyplot as plt

# Histograma de los datos
plt.hist(data, bins=30, density=True, alpha=0.5, label='Datos')

# Distribución normal ajustada
x_normal = np.linspace(0, 1, 100)
y_normal = norm.pdf(x_normal, mean_normal, std_normal)
plt.plot(x_normal, y_normal, 'r-', label='Distribución Normal')

# Distribución exponencial ajustada
x_exponential = np.linspace(0, 1, 100)
y_exponential = expon.pdf(x_exponential, scale_exponential)
plt.plot(x_exponential, y_exponential, 'g-', label='Distribución Exponencial')

plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.title('Histograma de datos y distribuciones ajustadas')
plt.legend()
plt.show()
```

4. Aparentemente ninguna de las 2 distribuciones se ajustan mucho a los datos, por lo menos visualmente.

```python
# 5.

from scipy.stats import kstest

# Prueba de Kolmogorov-Smirnov para la distribución normal
ks_normal = kstest(data, 'norm', args=(mean_normal, std_normal))

# Prueba de Kolmogorov-Smirnov para la distribución exponencial
ks_exponential = kstest(data, 'expon', args=(0, scale_exponential))

print("Prueba de Kolmogorov-Smirnov para la distribución normal: ", ks_normal)
print("Prueba de Kolmogorov-Smirnov para la distribución exponencial: ", ks_exponential)
```

5. Prueba de Kolmogorov-Smirnov para la distribución normal:
    Estadístico: El valor del estadístico es 0.0724. Este valor indica una máxima diferencia entre la función de distribución acumulada (CDF) de la distribución normal ajustada y la CDF de los datos empíricos. Cuanto menor sea el valor del estadístico, mejor será el ajuste de la distribución a los datos.
    Valor p: El valor p es 5.2664e-05, que es muy pequeño por lo que la distribución normal ajustada y la distribución empírica de los datos provengan de la misma distribución. Dado que el valor p es menor que el nivel de significancia 0.05, hay suficiente evidencia para rechazar la hipótesis nula y concluir que la distribución normal ajustada no se ajusta bien a los datos.

    Prueba de Kolmogorov-Smirnov para la distribución exponencial:
    Estadístico: El valor del estadístico es 0.1637. Igual que la anterior, cuanto menor sea el valor del estadístico, mejor será el ajuste de la distribución a los datos.
    Valor p : El valor p es 7.1714e-24, Ya que el valor p es mucho menor que el nivel de significancia 0.05, hay suficiente evidencia para rechazar la hipótesis nula y concluir que la distribución exponencial ajustada no se ajusta bien a los datos tampoco, apoyando de esta manera la respuesta del inciso 4.
